You have to scrape the job-title, job-location, company name, experience required.
The location filter to be used is ‚ÄúDelhi/NCR‚Äù. The salary filter to be used is ‚Äú3-6‚Äù lakhs
The task will be done as shown in the below steps:
1. first get the web page https://www.naukri.com/
2. Enter ‚ÄúData Scientist‚Äù in ‚ÄúSkill, Designations, and Companies‚Äù field.

4. Then apply the location filter and salary filter by checking the respective boxes

6. Finally create a dataframe of the scraped data. 

```python
driver.get("https://www.naukri.com/")
```

```python
designation = driver.find_element(By.CLASS_NAME,'suggestor-input')
designation.send_keys('Data Scientist')
```

```python
search = driver.find_element(By.CLASS_NAME,'qsbSubmit')
search.click()
```

```python
location_filter = driver.find_element(By.XPATH,'//*[@id="search-result-container"]/div[1]/div[1]/div/div/div[2]/div[5]/div[2]/div[2]/label')
location_filter.click()
```

```python
salary_filter = driver.find_element(By.XPATH,'//*[@id="search-result-container"]/div[1]/div[1]/div/div/div[2]/div[6]/div[2]/div[2]/label')
salary_filter.click()
```

```python
job_title = []
job_location = []
company_name = []
experience_req = []
salary_ = []
```

```python
title_tags = driver.find_elements(By.XPATH,'//div[@class="cust-job-tuple layout-wrapper lay-2 sjw__tuple "]/div/a')[0:11]
for i in title_tags:
    title = i.text
    job_title.append(title)
    
location_tags = driver.find_elements(By.XPATH,'//span[@class="ni-job-tuple-icon ni-job-tuple-icon-srp-location loc"]')[0:11]
for i in location_tags:
    location = i.text
    job_location.append(location)
    
company_tags = driver.find_elements(By.XPATH,'//a[@class=" comp-name mw-25"]')[0:11]
for i in company_tags:
    company = i.text
    company_name.append(company)
    
experience_tags = driver.find_elements(By.XPATH,'//span[@class="ni-job-tuple-icon ni-job-tuple-icon-srp-experience exp"]')[0:11]
for i in experience_tags:
    experience = i.text
    experience_req.append(experience)
    
salary_tags = driver.find_elements(By.XPATH,'//span[@class="ni-job-tuple-icon ni-job-tuple-icon-srp-rupee sal"]')[0:11]
for i in salary_tags:
    sr = i.text
    salary_.append(sr)
```

```python
print(len(job_title),len(job_location),len(company_name),len(experience_req),len(salary_))
```

```python
df = pd.DataFrame({'Title':job_title,'Location':job_location,'Company':company_name,'Experience':experience_req,'Salary':salary_})
df
```

```python
driver.get('https://www.flipkart.com/apple-iphone-11-black-64-gb/product-reviews/itm4e5041ba101fd?pid=MOBFWQ6BXGJCEYNY&lid=LSTMOBFWQ6BXGJCEYNYZXSHRJ&marketplace=FLIPKART')
```

```python
start = 0
end = 10
for page in range(start,end):
    rating_tags = driver.find_elements(By.XPATH,'//div[@class="XQDdHH Ga3i8K"]')[0:101]
    for i in rating_tags:
        ra = i.text
        rating.append(ra)
    
    review_tags = driver.find_elements(By.XPATH,'//p[@class="z9E0IG"]')[0:101]
    for i in review_tags:
        review = i.text
        review_summary.append(review)
    
    full_tags = driver.find_elements(By.XPATH,'//div[@class="ZmyHeo"]')[0:101]
    for i in full_tags:
        fullreview = i.text
        full_review.append(fullreview)
    
    next_button = driver.find_element(By.XPATH,'//a[@class="cn++Ap"]')
    next_button.click()
    time.sleep(10)
    
print(len(rating),len(review_summary),len(full_review))
```

```python
df = pd.DataFrame({'Rating':rating,'Review Summary':review_summary,'Full Review':full_review})
df
```

Q6: Write a python program to scrape data for Top 1000 Quotes of All Time.
The above task will be done in following steps:
1. First get the webpagehttps://www.azquotes.com/
2. Click on TopQuote
3. Than scrap a)Quote b) Author c) Type Of Quotes

```python
driver.get('https://www.azquotes.com/')
```

```python
quote = []
author = []
type_of_quote = []
```

```python
start = 1
end = 10

for page in range(start,end): 
    quote_tags = driver.find_elements(By.XPATH, '//a[@class="title"]')
    for i in quote_tags:
        Quote = i.text
        quote.append(Quote)
    
    author_tags = driver.find_elements(By.XPATH, '//div[@class="author"]')
    for i in author_tags:
        Author = i.text
        author.append(Author)
    
    type_of_quotes = driver.find_elements(By.XPATH, '//div[@class="tags"]')
    for i in type_of_quotes:
        Type = i.text
        type_of_quote.append(Type)
    
    next_button = driver.find_element(By.XPATH, '//li[@class="next"]/a')
    next_button.click()
    time.sleep(10)
    
```

```python
print(len(quote),len(author),len(type_of_quote))
```

```python
df = pd.DataFrame({'Quote':quote,'Author':author,'Type Of Quote':type_of_quote})
df.head(1000)
```
